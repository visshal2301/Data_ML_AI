LangChain defines several core message types (all subclasses of BaseMessage) to structure conversations with LLMs, distinguishing roles like user input, AI responses, and tools. These ensure type-safe handling of chat histories and multimodal content.

Core Message Types
HumanMessage: Captures user inputs or prompts, supporting text, images, or files.

SystemMessage: Provides instructions to prime the AI's behavior, tone, or context (e.g., "Respond concisely").

AIMessage: Holds the model's generated responses, including tool calls or metadata like token usage.

Additional Types
ToolMessage: Represents outputs from tool/function calls, feeding results back to the model.

AIMessageChunk: Partial AI responses during streaming (not for full messages).
​

FunctionMessage: Legacy type for older OpenAI function calling (superseded by ToolMessage).
​

Usage Notes
These types are imported from langchain_core.messages and passed as lists to chat models for multi-turn interactions. For example, a conversation might start with SystemMessage, add HumanMessage, then append AIMessage and ToolMessage as needed.

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from  langchain_openai import ChatOpenAI

llm_openai = ChatOpenAI(model="gpt-5-mini",temperature=0)
my_messages = [
    SystemMessage(content="You are a gen-z assistant who always answers in a fun way"),
    HumanMessage(content="Bro, tell me a fun fact")
]

llm_openai.invoke(my_messages).content
```

PromptTemplate in LangChain is a reusable structure for generating dynamic prompts to LLMs, using placeholders like {variable} that you fill with real values at runtime.

Key Components
Template string: Base text with placeholders, e.g., "Translate {text} to {language}."

Input variables: List of expected placeholders, e.g., ["text", "language"].

Values: Data passed to .format(**values) to create the final prompt.

```python
from langchain_core.prompts import PromptTemplate


user_input = input("Enter a topic for fun fact : ")

dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic" : user_input})

llm_openai.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate 

prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are a {tone} assistant"),
    ("user","Write a fun fact about {topic}.")
])

user_input = input("Enter a topic: ")
user_tone = input("Enter a tone: ")

ready_prompt = prompt_template.invoke({"topic":user_input,"tone":user_tone})

llm_openai.invoke(ready_prompt.messages).content
```
